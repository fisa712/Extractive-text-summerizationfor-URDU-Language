### Pytorch word embeddings
* Genrating word embeddings of each senetence of the document.


```python
# importing pytorh_libraries

import torch
import torch.nn as nn
from torch.nn.functional import cosine_similarity

torch.manual_seed(1)
import torch.nn.functional as F

import torch.optim as optim
import torch
```

#### Claculating bigrams
  * embedding dimentions =  100

```python
bigrams = 2
embed_dim= 100

#genrating n-grams

ngrams = [
    (
        [processed_text1[i - j - 1] for j in range(bigrams)], processed_text1[i]
    )
                    for i in range(bigrams, len(processed_text1))
         ]
#print few ngrams(bigrams)
# print(ngrams[:10])

vocabulary = set(processed_text1)

#create a dictionary

word_to_dict = {
    
    word: i for i, word in enumerate(vocabulary)
    
                }

   

print(ngrams[0:5])
```

### Pytorch 
* word embeddings 
    * genrating embedding vectors of Tensors

```python

class NGramLanguageModeler(nn.Module):

    def __init__(self, vocab_size, embedding_dim, context_size):
        super(NGramLanguageModeler, self).__init__()
        self.embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.linear1 = nn.Linear(context_size * embedding_dim, 128)
        self.linear2 = nn.Linear(128, vocab_size)
        self.linear3 = nn.Linear(context_size * embedding_dim, 128)
        self.linear2 = nn.Linear(128, vocab_size)
        
        
    def forward(self, inputs):
        embeds = self.embeddings(inputs).view((1, -1))
        out = F.relu(self.linear1(embeds))
        out = self.linear2(out)
        softmax_log_probs = F.log_softmax(out, dim=1)
        return softmax_log_probs


    
# calculating losses

losses = []
loss_function = nn.NLLLoss()
embd_model = NGramLanguageModeler(len(vocabulary), embed_dim, bigrams)
optimizer = optim.SGD(embd_model.parameters(), lr=0.001)

# for almost 10 epochs

for epoch in range(10):
    tot_loss = 0
    for context, target in ngrams:

    #   inputs of model ->genrate tensors of int indics from the words 
        
        context_idxs = torch.tensor([word_to_dict[w] for w in context], dtype=torch.long)

        
    # zero_gradients bef passing to the new instnc from old ones
        
        embd_model.zero_grad()

        
    #    calculating  log probabilities
    
        softmax_log_probs = embd_model(context_idxs)
   
    # traget word -> in a tensor 
    # loss function. 
        loss = loss_function(softmax_log_probs, torch.tensor([word_to_dict[target]], dtype=torch.long))

    # backward pass and 
        loss.backward()
        
        optimizer.step()

        # total_loss_> by tensor.item()
        tot_loss += loss.item()
    losses.append(tot_loss)
#printi to check for any word
# print(model.embeddings.weight[word_to_dict["اموات"]])



# getting embeddings of each word
embeded_vec = {}
for key , _ in word_to_dict.items():
#     print(key , " -> " , _ )
    embeded_vec[key] = (embd_model.embeddings.weight[word_to_dict[key]])
print(embeded_vec)
```

### Similarity matrix 
* Cosine Similarity Matrix
    * if duistanec is equal and near to 1 closely related 
    * otherwise worst case would be assigned -1

```python
from sklearn.metrics.pairwise import cosine_similarity
#genrating a 2-d matrix for similarity matrix
sim_mat = np.zeros([len(procesed_text), len(procesed_text)])
```

```python
for i in range(len(procesed_text)):
    for j in range(len(procesed_text)):
        if i != j:
              sim_mat[i][j] = cosine_similarity(tvect[i].reshape(1,100), tvect[j].reshape(1,100))[0,0]
```

### Text-rank algorithm
* Inspired By page-rank algorithm
* Using a graph implimented using SImilarity matrix

```python
import networkx as nx
nx_graph = nx.from_numpy_array(sim_mat)
# scores = nx.pagerank(nx_graph , max_iter=100 ,  tol=1.0e-3)
# G = nx.DiGraph(nx.path_graph(4))
pr = nx.pagerank_numpy(nx_graph, alpha=0.8)
```

```python
ranked_sentences = sorted(((pr[i],s) for i,s in enumerate(procesed_text)), reverse=True)
```
